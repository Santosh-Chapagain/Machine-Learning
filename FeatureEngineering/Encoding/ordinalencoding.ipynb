```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('customer_data.csv')
```

```python
df
```

```python
def clean_data(df):
    # Drop column: 'age'
    df = df.drop(columns=['age'])
    # Drop column: 'id'
    df = df.drop(columns=['id'])
    # Drop column: 'gender'
    df = df.drop(columns=['gender'])
    # Drop column: 'income'
    df = df.drop(columns=['income'])
    # Drop column: 'region'
    df = df.drop(columns=['region'])
    # Drop column: 'purchase_amount'
    df = df.drop(columns=['purchase_amount'])
    # Drop column: 'product_category'
    df = df.drop(columns=['product_category'])
    # Drop column: 'promotion_usage'
    df = df.drop(columns=['promotion_usage'])
    # Drop column: 'satisfaction_score'
    df = df.drop(columns=['satisfaction_score'])
    return df

data = clean_data(df.copy())
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_test , x_train , y_test , y_train = train_test_split(data.iloc[:,0:2] , data.iloc[: , -1] , test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['HighSchool', 'College', 'Bachelor', 'Masters'], ['Regular', 'Silver', 'Gold', ]])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```
