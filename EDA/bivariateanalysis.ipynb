```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
```

### 1. Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex',style='smoker',size='size')
```

### 2. Barplot (Numerical-Categorical)

```python
sns.barplot(data=titanic, x='pclass', y='age',hue='sex')
```

### 3. Boxplot(Numerical-Categorical)

```python
sns.boxplot(data=titanic, x='sex', y='age',hue='survived')
```

### 4. Kdeplot(Numerical-Categorical)

```python
sns.kdeplot(titanic[titanic['survived']==0]['age'].dropna())
sns.kdeplot(titanic[titanic['survived']==1]['age'].dropna())
plt.show()
```

### 5. HeatMap(Categorical-Categorical)

```python
pd.crosstab(titanic['pclass'], titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()*100
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python

ct = pd.crosstab(titanic['sibsp'], titanic['survived'])
print(ct)
sns.clustermap(ct, figsize=(5, 5))
plt.show()
```

### 6. Pairplot

```python
sns.pairplot(iris, hue='species')
```
